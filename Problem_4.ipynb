# Problem 4: Working with the Iris Dataset


```python
# Task 4.1: Read and Clean Data
# -----------------------------
# Import pandas to handle the dataset
import pandas as pd

# Load the Iris dataset from the provided CSV file
iris_df = pd.read_csv("iris.csv")

# Display the first few rows of the dataset to confirm successful loading
print("First few rows of the Iris dataset:")
print(iris_df.head())

# Check the column names to ensure they are correct
print("\nColumn Names in iris_df:", iris_df.columns)

# Rename columns if they do not match the expected names
expected_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
if list(iris_df.columns) != expected_columns:
    iris_df.columns = expected_columns
    print("\nColumns have been renamed to match the expected format.")
else:
    print("\nColumn names are as expected.")

# Check for missing values in each column
missing_values = iris_df.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values)
```

```python
# Task 4.2: Explore and Describe the Data
# ---------------------------------------
# Display summary statistics for each numeric column (mean, min, max)
iris_summary = iris_df.describe().loc[['mean', 'min', 'max']]
print("\nSummary Statistics (Mean, Min, Max) for Each Numeric Column:")
print(iris_summary)

# Calculate the median for each numeric column separately
iris_median = iris_df.median(numeric_only=True)
print("\nMedian Values for Each Numeric Column:")
print(iris_median)

# Count the number of unique species in the dataset
species_count = iris_df['species'].value_counts()
print("\nSpecies Count:")
print(species_count)
```

```python
# Task 4.3: Visualize the Data
# ----------------------------
# Import seaborn and matplotlib for data visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure column names are correct in case of prior issues
expected_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
if list(iris_df.columns) != expected_columns:
    iris_df.columns = expected_columns
    print("Columns have been renamed to match the expected format.")
else:
    print("Column names are as expected.")

# Set the style for the plot
sns.set(style="ticks")

# Create a pair plot with seaborn, coloring by species
pair_plot = sns.pairplot(iris_df, hue="species", markers=["o", "s", "D"], palette="Set2")
pair_plot.fig.suptitle("Pair Plot of Iris Dataset Features", y=1.02)  # Adjust title position

# Show the plot
plt.show()
```
